```
#Начинаем все с импорта необходимых модулей
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Подгружаем изображение 
!wget "https://memepedia.ru/wp-content/uploads/2018/05/cover-1-27-768x551.jpg" -O dt.png
image = cv2.imread('./dt.png')
# Включаем нормальный цвет
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

h, w = image.shape[:2]
tx = 100
ty = 200
M = np.float32([[1, 0, tx], [0, 1, ty]])  # tx - смещение по горизонтали
M_rotate = cv2.getRotationMatrix2D((w//2,h//2),45,0.5)
```

```
image_translated = cv2.warpAffine(image, M, (w,h))
image_rotated = cv2.warpAffine(image, M_rotate, (w,h))
image_transposed = cv2.transpose(image)
image_flipped = cv2.flip(image,1) # vertcal 1 / horizontal 0
image_resized = cv2.resize(image, None, fx = 0.75, fy = 0.75)
image_resized_2 = cv2.resize(image, (800,600), interpolation = cv2.INTER_AREA)

plt.imshow(image)
plt.show()
plt.imshow(image_translated)
plt.show()
plt.imshow(image_rotated)
plt.show()
plt.imshow(image_transposed)
plt.show()
```

Маскирование и побитовые операции

```
square = np.zeros((300,300),dtype= np.uint8)
cv2.rectangle(square,(50,50),(250,250),255,-1)
ellipse = np.zeros((300,300),dtype=np.uint8)
cv2.ellipse(ellipse,(150,150),(150,150),30,0,180,255,-1)

# В colab нужно поменять цвет
square = cv2.cvtColor(square, cv2.COLOR_BGR2RGB)
ellipse = cv2.cvtColor(ellipse, cv2.COLOR_BGR2RGB)


plt.imshow(square)
plt.show()

plt.imshow(ellipse)
plt.show()
```

```
image_and = cv2.bitwise_and(square,ellipse)
plt.imshow(image_and)
plt.show()
```

```
image_or = cv2.bitwise_or(square,ellipse)
plt.imshow(image_or)
plt.show()
```

```
image_xor = cv2.bitwise_xor(square,ellipse)
plt.imshow(image_xor)
plt.show()
```

```
image_not_square = cv2.bitwise_not(square)
plt.imshow(image_not_square)
plt.show()

image_not_ellipse = cv2.bitwise_not(ellipse)
plt.imshow(image_not_ellipse)
plt.show()
```
