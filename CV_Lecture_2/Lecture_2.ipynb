```
#Начинаем все с импорта необходимых модулей
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = np.zeros((600, 800, 3), dtype = np.uint8) #(height , width, кол-во каналов в 3 измерении)
h,w,ch = img.shape#[:2]  # будем использовать по умолчанию, чтобы принебречь ошибки
print(h,w,ch)
```

Изменем значения pixel внутри изображения

```
for py in range(0, h):
  for px in range(w):
    img[py,px]=(0,0,255)

# На питоне будет работать нормально, но тут крашнется
#cv2.imshow('black',img)
#cv2.waitKey()
#cv2.destroyAllWindows()

# Поэтому заменим это дело на:
plt.imshow(img)
plt.show()
```

Функции

```
img_1 = np.ones((600,800,3), dtype=np.uint8)*255

# Формируем случайный цвет
def randColor():
  c = np.random.uniform(0, 255, 3)
  return (int(c[0]), int(c[1]), int(c[2]))

# Формируем случайные координаты
def randCoords(image):
  height, width = image.shape[:2]
  return (np.random.randint(0, width),np.random.randint(0, height))
# Рисуем
def drawLines(image, iter):
  for i in range (iter):
    cv2.line(image, randCoords(image), randCoords(image), randColor())

def drawTriangles(image, iter):
  verticles=np.zeros((3,2),np.int)
  for i in range (iter):
    for j in range (3):
      verticles[j,0] = randCoords(image)[0]
      verticles[j,1] = randCoords(image)[1]
    cv2.fillPoly(image,[verticles],randColor())

def drawCircles(image, iter):
  for i in range (iter):
    cv2.circle(image, randCoords(image), np.random.randint(20, 170), randColor(), np.random.randint(-15, 15)




#drawLines(img_1, 100)
drawTriangles(img_1,100)
drawCircles(img_1,150)

#print(randColor())
#print(randCoords(img_1))
```
